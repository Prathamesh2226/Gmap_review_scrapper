```python
from textblob import TextBlob
import pandas as pd
# Function to analyze sentiment and polarity
def analyze_sentiment_with_score(review):
    polarity = TextBlob(review).sentiment.polarity
    if polarity > 0:
        sentiment = 'Positive'
    elif polarity < 0:
        sentiment = 'Negative'
    else:
        sentiment = 'Neutral'
    return polarity, sentiment

# Apply the function and add two new columns
df[['Polarity', 'Sentiment']] = df['Review'].apply(lambda x: pd.Series(analyze_sentiment_with_score(x)))
df
```

```python

# Function to analyze sentiment and polarity
def analyze_sentiment_with_score(review):
    try:
        polarity = TextBlob(review).sentiment.polarity
        if polarity > 0:
            sentiment = 'Positive'
        elif polarity < 0:
            sentiment = 'Negative'
        else:
            sentiment = 'Neutral'
        return polarity, sentiment
    except Exception as e:  # Catch potential errors during sentiment analysis
        print(f"Error analyzing sentiment for review: {review} - Error: {e}")
        return 0, 'Neutral'  # Return neutral sentiment in case of errors


# Apply the function and add two new columns
df[['Polarity', 'Sentiment']] = df['Review'].apply(lambda x: pd.Series(analyze_sentiment_with_score(x)))


# Create the pie chart
sentiment_counts = df['Sentiment'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(sentiment_counts, labels=sentiment_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Sentiment Distribution')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python

positive_reviews = df[df['Sentiment'] == 'Positive']
negative_reviews = df[df['Sentiment'] == 'Negative']

positive_percentage = (len(positive_reviews) / len(df)) * 100
negative_percentage = (len(negative_reviews) / len(df)) * 100

print(f"Positive Reviews: {positive_percentage:.2f}%")
print(f"Negative Reviews: {negative_percentage:.2f}%")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

# Filter negative reviews
negative_reviews_text = df[df['Sentiment'] == 'Negative']['Review']

# Use TF-IDF to find important terms in negative reviews
vectorizer = TfidfVectorizer(stop_words='english', ngram_range=(1, 2), max_features=10)
X = vectorizer.fit_transform(negative_reviews_text)

# Get the terms/phrases
terms = vectorizer.get_feature_names_out()

# Convert to a DataFrame for easy readability
tfidf_df = pd.DataFrame(X.toarray(), columns=terms)

# Sum the TF-IDF scores for each term across all negative reviews
term_scores = tfidf_df.sum(axis=0).sort_values(ascending=False)
term_scores
```
